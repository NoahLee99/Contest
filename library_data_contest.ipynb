## 3. 데이터프레임 전처리 및 병합

```python
## df_age_gender(2021~2023 데이터) 전처리 및 Long 형식 변환

# 첫 번째 행(메타데이터) 제거
df_age_gender_processed = df_age_gender.iloc[1:].copy()

# 필요한 컬럼 선택 및 이름 변경 (컬럼명에서 1 → 남자, 2 → 여자)
df_age_gender_melted = df_age_gender_processed[['행정구역별(읍면동)', '연령별', '2021.1', '2021.2', '2022.1', '2022.2', '2023.1', '2023.2']].copy()

# '행정구역별(읍면동)' 컬럼 이름을 '지역'으로 변경
df_age_gender_melted = df_age_gender_melted.rename(columns={'행정구역별(읍면동)': '지역'})


# melt 함수를 각각 사용하여 각각 남자 총인구와 여자 총인구 데이터를 Long 형식으로 변환
# (long 형식으로 연도와 성별을 행으로 풀면 시각화시에 유용함)
df_age_gender_male_melted = df_age_gender_melted.melt(id_vars=['지역', '연령별'],
                                                     value_vars=['2021.1', '2022.1', '2023.1'],  # 녹일(melt) 컬럼들
                                                     var_name='연도_남자',
                                                     value_name='총인구_남자(명)')

df_age_gender_female_melted = df_age_gender_melted.melt(id_vars=['지역', '연령별'],
                                                       value_vars=['2021.2', '2022.2', '2023.2'],
                                                       var_name='연도_여자',
                                                       value_name='총인구_여자(명)')

# '연도_남자' 및 '연도_여자' 컬럼 이름에서 확장자를 제거하여 '연도' 컬럼과 일치시키기
df_age_gender_male_melted['연도'] = df_age_gender_male_melted['연도_남자'].str.replace('.1', '', regex=False)
df_age_gender_male_melted = df_age_gender_male_melted.drop(columns=['연도_남자'])

df_age_gender_female_melted['연도'] = df_age_gender_female_melted['연도_여자'].str.replace('.2', '', regex=False)
df_age_gender_female_melted = df_age_gender_female_melted.drop(columns=['연도_여자'])


# 남자 총인구와 여자 총인구 데이터프레임을 '지역', '연령별', '연도'를 기준으로 병합
df_age_gender_combined = pd.merge(df_age_gender_male_melted, df_age_gender_female_melted, on=['지역', '연령별', '연도'], how='outer')

# 총인구(명) 계산
df_age_gender_combined['총인구(명)'] = (
    pd.to_numeric(df_age_gender_combined['총인구_남자(명)'], errors='coerce')
    + pd.to_numeric(df_age_gender_combined['총인구_여자(명)'], errors='coerce')
)

df_age_gender_combined.head()
```

```python
## df_age_gender_past(2015~2020 데이터) 전처리 및 Long 형식 변환

# 첫 번째 행 제거
df_age_gender_past_processed = df_age_gender_past.iloc[1:].copy()

# 필요한 컬럼 선택 및 이름 변경
df_age_gender_past_melted = df_age_gender_past_processed[['행정구역별(읍면동)', '연령별',
                                                          '2015.1', '2015.2',
                                                          '2016.1', '2016.2',
                                                          '2017.1', '2017.2',
                                                          '2018.1', '2018.2',
                                                          '2019.1', '2019.2',
                                                          '2020.1', '2020.2']].copy()

# '행정구역별(읍면동)' 컬럼 이름을 '지역'으로 변경
df_age_gender_past_melted = df_age_gender_past_melted.rename(columns={'행정구역별(읍면동)': '지역'})

# melt 함수를 각각 사용하여 각각 남자 총인구와 여자 총인구 데이터를 Long 형식으로 변환
df_age_gender_past_male_melted = df_age_gender_past_melted.melt(id_vars=['지역', '연령별'],
                                                     value_vars=['2015.1', '2016.1', '2017.1',
                                                                 '2018.1', '2019.1', '2020.1'],
                                                     var_name='연도_남자',
                                                     value_name='총인구_남자(명)')

df_age_gender_past_female_melted = df_age_gender_past_melted.melt(id_vars=['지역', '연령별'],
                                                       value_vars=['2015.2', '2016.2', '2017.2',
                                                                   '2018.2', '2019.2', '2020.2'],
                                                       var_name='연도_여자',
                                                       value_name='총인구_여자(명)')

# '연도_남자' 및 '연도_여자' 컬럼 이름에서 확장자 제거하여 '연도' 컬럼과 일치시키기
df_age_gender_past_male_melted['연도'] = df_age_gender_past_male_melted['연도_남자'].str.replace('.1', '', regex=False)
df_age_gender_past_male_melted = df_age_gender_past_male_melted.drop(columns=['연도_남자'])

df_age_gender_past_female_melted['연도'] = df_age_gender_past_female_melted['연도_여자'].str.replace('.2', '', regex=False)
df_age_gender_past_female_melted = df_age_gender_past_female_melted.drop(columns=['연도_여자'])


# 남자 총인구와 여자 총인구 데이터프레임을 '지역', '연령별', '연도'를 기준으로 병합
df_age_gender_past_combined = pd.merge(df_age_gender_past_male_melted, df_age_gender_past_female_melted, on=['지역', '연령별', '연도'], how='outer')

# 총인구(명) 계산
df_age_gender_past_combined['총인구(명)'] = pd.to_numeric(df_age_gender_past_combined['총인구_남자(명)'], errors='coerce') + pd.to_numeric(df_age_gender_past_combined['총인구_여자(명)'], errors='coerce')


df_age_gender_past_combined.head()
```

```python
## df_pop_trend 전처리 및 Long 형식 변환 (2021-2023 데이터)

# 첫 번째 행을 컬럼 설명으로 사용하기 위해 임시 보존
df_pop_trend_header_2123 = df_pop_trend.iloc[0].copy()

# 첫 번째 행 제거
df_pop_trend_processed_2123 = df_pop_trend.iloc[1:].copy()

# '행정구역별(1)' 컬럼명을 '지역'으로 변경
df_pop_trend_processed_2123 = df_pop_trend_processed_2123.rename(columns={'행정구역별(1)': '지역'})

# 2021, 2022, 2023년 컬럼 목록 생성
years_trend_2123 = ['2021', '2022', '2023']
value_vars_trend_2123 = [col for year in years_trend_2123 for col in [year] + [f'{year}.{i}' for i in range(1, 5)]]

# '지역' 컬럼과 선택된 연도별 동향 컬럼들만 선택
df_pop_trend_melted_2123 = df_pop_trend_processed_2123[['지역'] + value_vars_trend_2123]

# 원본 컬럼 이름과 세부 항목 매핑
pop_trend_col_map_2123 = df_pop_trend_header_2123[value_vars_trend_2123].to_dict()

# Long 형식으로 변환
df_pop_trend_melted_2123 = df_pop_trend_melted_2123.melt(
    id_vars=['지역'],
    value_vars=value_vars_trend_2123,
    var_name='원본컬럼_동향',
    value_name='인구동향값' # 컬럼명 변경
)

# '원본컬럼_동향'을 사용하여 '연도'와 '세부항목_동향' 컬럼 생성
df_pop_trend_melted_2123['연도'] = df_pop_trend_melted_2123['원본컬럼_동향'].str.extract(r'(\d{4})')
# 원본 컬럼 이름을 사용하여 매핑
df_pop_trend_melted_2123['세부항목_동향'] = df_pop_trend_melted_2123['원본컬럼_동향'].map(pop_trend_col_map_2123)


# 불필요한 '원본컬럼_동향' 컬럼 삭제
df_pop_trend_melted_2123 = df_pop_trend_melted_2123.drop(columns=['원본컬럼_동향'])

# '인구동향값' 컬럼의 데이터 타입을 숫자로 변환
df_pop_trend_melted_2123['인구동향값'] = pd.to_numeric(df_pop_trend_melted_2123['인구동향값'], errors='coerce')

# 결측치 채우기
df_pop_trend_melted_2123['인구동향값'] = df_pop_trend_melted_2123['인구동향값'].fillna(0)

# 데이터프레임 확인
display(df_pop_trend_melted_2123.head())


## df_pop_trend_past 전처리 (2015-2020 데이터)

# 첫 번째 행을 컬럼 설명으로 사용하기 위해 임시 보존
df_pop_trend_header_1520 = df_pop_trend_past.iloc[0].copy()

# 첫 번째 행 제거
df_pop_trend_past_processed = df_pop_trend_past.iloc[1:].copy()

# '행정구역별(1)' 컬럼명을 '지역'으로 변경
df_pop_trend_past_processed = df_pop_trend_past_processed.rename(columns={'행정구역별(1)': '지역'})

# 2015-2020년 컬럼 목록 생성
years_trend_1520 = [str(y) for y in range(2015, 2021)]
value_vars_trend_1520 = [col for year in years_trend_1520 for col in [year] + [f'{year}.{i}' for i in range(1, 5)]]

# '지역' 컬럼과 선택된 연도별 동향 컬럼들만 선택
df_pop_trend_melted_1520 = df_pop_trend_past_processed[['지역'] + value_vars_trend_1520]

# 원본 컬럼 이름과 세부 항목 매핑
pop_trend_col_map_1520 = df_pop_trend_header_1520[value_vars_trend_1520].to_dict()

# Long 형식으로 변환
df_pop_trend_melted_1520 = df_pop_trend_melted_1520.melt(
    id_vars=['지역'],
    value_vars=value_vars_trend_1520,
    var_name='원본컬럼_동향',
    value_name='인구동향값' # 컬럼명 변경
)

# '원본컬럼_동향'을 사용하여 '연도'와 '세부항목_동향' 컬럼 생성
df_pop_trend_melted_1520['연도'] = df_pop_trend_melted_1520['원본컬럼_동향'].str.extract(r'(\d{4})')
# 원본 컬럼 이름을 사용하여 매핑
df_pop_trend_melted_1520['세부항목_동향'] = df_pop_trend_melted_1520['원본컬럼_동향'].map(pop_trend_col_map_1520)


# 불필요한 '원본컬럼_동향' 컬럼 삭제
df_pop_trend_melted_1520 = df_pop_trend_melted_1520.drop(columns=['원본컬럼_동향'])

# '인구동향값' 컬럼의 데이터 타입을 숫자로 변환
df_pop_trend_melted_1520['인구동향값'] = pd.to_numeric(df_pop_trend_melted_1520['인구동향값'], errors='coerce')

# 결측치 채우기
df_pop_trend_melted_1520['인구동향값'] = df_pop_trend_melted_1520['인구동향값'].fillna(0)

# 데이터프레임 확인
display(df_pop_trend_melted_1520.head())


# 두 데이터프레임을 세로(행) 방향으로 이어 붙이기
merged_df_pop_trend_all_years = pd.concat([df_pop_trend_melted_1520, df_pop_trend_melted_2123], ignore_index=True)

# 합쳐진 데이터프레임을 '지역'과 '연도', '세부항목_동향' 컬럼을 기준으로 정렬
merged_df_pop_trend_all_years_sorted = merged_df_pop_trend_all_years.sort_values(by=['지역', '연도', '세부항목_동향']).reset_index(drop=True)

# 정렬된 데이터프레임 정보 확인
merged_df_pop_trend_all_years_sorted.info()

# 정렬된 데이터프레임 확인
display(merged_df_pop_trend_all_years_sorted.head())
display(merged_df_pop_trend_all_years_sorted.tail())
```

```python
# age_gender 데이터프레임의 '총인구(명)', '총인구_남자(명)', '총인구_여자(명)' 컬럼을 정수형으로 변환
merged_df_age_gender_all_years['총인구(명)'] = pd.to_numeric(merged_df_age_gender_all_years['총인구(명)'], errors='coerce').fillna(0).astype(int)
merged_df_age_gender_all_years['총인구_남자(명)'] = pd.to_numeric(merged_df_age_gender_all_years['총인구_남자(명)'], errors='coerce').fillna(0).astype(int)
merged_df_age_gender_all_years['총인구_여자(명)'] = pd.to_numeric(merged_df_age_gender_all_years['총인구_여자(명)'], errors='coerce').fillna(0).astype(int)


## 최종 병합을 위한 첫 번째 age_gender 데이터프레임
final_merged_df = merged_df_age_gender_all_years.copy()


## household 데이터프레임 병합
# '세대구성별' 및 '세부항목' 컬럼을 pivot하여 병합 준비
merged_df_household_pivot = merged_df_household_all_years_sorted.pivot_table(
    index=['지역', '연도'],
    columns=['세대구성별', '세부항목'],
    values='가구수',
    aggfunc='sum'
).reset_index()

# 컬럼 이름 정리
merged_df_household_pivot.columns = [f'{col[0]}_{col[1]}' if col[0] != '지역' and col[0] != '연도' else col[0] for col in merged_df_household_pivot.columns]

final_merged_df = pd.merge(final_merged_df, merged_df_household_pivot, on=['지역', '연도'], how='left')


## pop_dens 데이터프레임 병합
final_merged_df = pd.merge(final_merged_df, merged_df_pop_dens_all_years_sorted, on=['지역', '연도'], how='left')


## job 데이터프레임 병합
# '산업분류' 및 '규모별' 컬럼을 pivot하여 병합 준비
merged_df_job_pivot = merged_df_job_all_years_sorted.pivot_table(
    index=['지역', '연도'],
    columns=['산업분류', '규모별'],
    values='총종사자수(명)',
    aggfunc='sum'
).reset_index()

# 컬럼 이름 정리
merged_df_job_pivot.columns = [f'{col[0]}_{col[1]}' if col[0] != '지역' and col[0] != '연도' else col[0] for col in merged_df_job_pivot.columns]

final_merged_df = pd.merge(final_merged_df, merged_df_job_pivot, on=['지역', '연도'], how='left')


## pop_trend 데이터프레임 병합
# '세부항목_동향' 컬럼을 pivot하여 병합 준비
merged_df_pop_trend_pivot = merged_df_pop_trend_all_years_sorted.pivot_table(
    index=['지역', '연도'],
    columns='세부항목_동향',
    values='인구동향값',
    aggfunc='sum'
).reset_index()

# 컬럼 이름 정리
merged_df_pop_trend_pivot.columns.name = None

final_merged_df = pd.merge(final_merged_df, merged_df_pop_trend_pivot, on=['지역', '연도'], how='left')


# 최종 병합된 데이터프레임 정보 확인
final_merged_df.info()

# 최종 병합된 데이터프레임 확인
display(final_merged_df.head())
display(final_merged_df.tail())
```

```python
## 최종 final_merged_df 전처리

# 제거할 총계/요약 값 목록 정의
summary_age_values = ['합계', '평균연령', '중위연령']
summary_region_values = ['전국']
summary_household_values = ['계']
summary_job_values = ['전체 산업', '전체']
summary_scale_values = ['계', '전규모']

# '연령별' 컬럼에서 총계 값 제거
final_merged_df_filtered = final_merged_df[~final_merged_df['연령별'].isin(summary_age_values)].copy()

# '지역' 컬럼에서 '전국' 제거
final_merged_df_filtered = final_merged_df_filtered[~final_merged_df_filtered['지역'].isin(summary_region_values)].copy()

# 필터링된 데이터프레임의 정보 및 처음 다섯 행 확인
final_merged_df_filtered.info()
display(final_merged_df_filtered.head())
```
